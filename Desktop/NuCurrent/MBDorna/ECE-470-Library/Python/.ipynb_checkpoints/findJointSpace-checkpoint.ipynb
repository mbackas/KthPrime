```python
from ece470_lib import *

a0 = [0,0,1];
a1 = [0,1,0];
a2 = [0,1,0];
a3 = [0,1,0];
a4 = [1,0,0];
a5 = [1,0,0];

inTOmm = 25.4;

q0 = [0,0,0];
q1 = [3.76*inTOmm,0,8.11*inTOmm];
q2 = [11.76*inTOmm,0,8.11*inTOmm];
q3 = [17.76*inTOmm,0,8.11*inTOmm];
q4 = [19.48*inTOmm,0,8.11*inTOmm];
q5 = [19.48*inTOmm,50,8.11*inTOmm];

s0 = toScrew(a0,q0);
s1 = toScrew(a1,q1);
s2 = toScrew(a2,q2);
s3 = toScrew(a3,q3);
s4 = toScrew(a4,q4);
s5 = toScrew(a5,q5);

S=[s0,s1,s2,s3,s4]#,s5]

toolRot = np.array([[ 0,0,1],
                    [ 0,1,0],
                    [-1,0,0]]);
toolPos = np.array([[19.48*inTOmm+600 , 0 , 8.11*inTOmm]]).transpose();
M=toPose(toolRot,toolPos);

goalRot = np.eye(3);
X=np.arange(-10,480,5);
# Y=X;
Y=np.array([0])
Z=np.arange(600,1225,5)
validTheta = np.zeros((X.size, Y.size, Z.size));
thetaMap = [[[np.ones((len(S[0]),1))*-99 for i in range(Z.size)] for j in range(Y.size)] for k in range(X.size)];
valCount=0
invalCount = 0
totCount=0
lastValTheta = np.zeros((len(S[0]),1))
for i in range(X.size):
    for j in range(Y.size):
        for k in range(Z.size):
            print("valid=",valCount,"invalid=",invalCount,"out of",totCount,X.size*Y.size*Z.size,[X[i],Y[j],Z[k]])
            totCount+=1
            goalPos = np.array([[X[i],Y[j],Z[k]]]).transpose();
            goalT = toPose(goalRot,goalPos);
            theta,norm = findIK(goalT, S,M,theta=lastValTheta,max_iter=100);
            if(norm<0.01):
                valCount+=1
                for p in range(theta.size):
                    while(theta[p]>np.pi):
                        theta[p]-=2*np.pi
                    while(theta[p]<-np.pi):
                        theta[p]+=2*np.pi
                lastValTheta=theta
                validTheta[i,j,k]=1;
                thetaMap[i][j][k]=theta;
            else:
                invalCount+=1
                print(norm)
```
